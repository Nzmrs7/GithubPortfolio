```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("ifood_df.csv")
df.columns = df.columns.str.replace(' ', '')
df.T
```

1. Create marital status column and combine marital_Divorced, marital_Married, marital_Single, marital_Together
2. Combine 	education_2n Cycle and education_Master as education_postgrad
3. Create education column and combine education_Basic, education_Graduation, education_Master, education_PhD


```python
df['education_Master'] = df['education_Master'] + df['education_2nCycle']
df['Dependent'] = df["Kidhome"] + df["Teenhome"]
```

```python
df['MaritalStatus'] = 0
```

```python
df.loc[df['marital_Divorced'] == 1, 'MaritalStatus'] = 'Divorced'
df.loc[df['marital_Married'] == 1, 'MaritalStatus'] = 'Married'
df.loc[df['marital_Single'] == 1, 'MaritalStatus'] = 'Single'
df.loc[df['marital_Together'] == 1, 'MaritalStatus'] = 'Together'
df.loc[df['marital_Widow'] == 1, 'MaritalStatus'] = 'Widow'
```

```python
df.loc[df['AcceptedCmp1'] == 1, 'AcceptedCmp1'] = 'Yes'
df.loc[df['AcceptedCmp1'] == 0, 'AcceptedCmp1'] = 'No'
df.loc[df['AcceptedCmp2'] == 1, 'AcceptedCmp2'] = 'Yes'
df.loc[df['AcceptedCmp2'] == 0, 'AcceptedCmp2'] = 'No'
df.loc[df['AcceptedCmp3'] == 1, 'AcceptedCmp3'] = 'Yes'
df.loc[df['AcceptedCmp3'] == 0, 'AcceptedCmp3'] = 'No'
df.loc[df['AcceptedCmp4'] == 1, 'AcceptedCmp4'] = 'Yes'
df.loc[df['AcceptedCmp4'] == 0, 'AcceptedCmp4'] = 'No'
df.loc[df['AcceptedCmp5'] == 1, 'AcceptedCmp5'] = 'Yes'
df.loc[df['AcceptedCmp5'] == 0, 'AcceptedCmp5'] = 'No'
```

```python
df['Education'] = 0
df.loc[df['education_Basic'] == 1, 'Education'] = 'High School'
df.loc[df['education_Graduation'] == 1, 'Education'] = 'Bachelor'
df.loc[df['education_Master'] == 1, 'Education'] = 'Master'
df.loc[df['education_PhD'] == 1, 'Education'] = 'PhD'
```

```python
df.drop(['education_Basic','education_2nCycle', 'education_Graduation','education_Master','education_PhD', "Kidhome","Teenhome",'marital_Divorced','marital_Married','marital_Single','marital_Together','marital_Widow'], axis=1, inplace=True)
df.T
```

```python
df['Dependent'] = df['Dependent'].apply(str)
df['AcceptedCmpOverall'] = df['AcceptedCmpOverall'].apply(str)
```
